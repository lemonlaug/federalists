Comparing Classification Algorithms with the Federalist Papers
====================================
The Federalist papers are a famous classification problem. Most of the Federalist papers are directly attributed to Alexander Hamilton or James Madison (a small number are written by John Jay), but 12 of the papers were published without attribution, and while it is agreed their author is either Hamilton or Madison, at one point, the exact attribution was a subject of some debate. Although the debate is considered settled today, it hasn't diminished the popularity of these documents as an example for doing classification based on text.

In that spirit, I was thinking about the federalist papers recently as I looked for some data to brush up on classification algorithms. Oftentimes in work or school, one thinks about a single algorithm and its properties in depth, without necessarily comparing it to other algorithms, so I thought it would be a helpful refresher to examine some common classifiers side-by-side, to illustrate some of their differences.

The Federalist Papers are all available from the Library of Congress, so we'll begin with raw data and download them from that source.

```python
#Get the data from the library of congress.
#Begin by downloading the index of papers, so we can get the link for each paper.
base = 'http://thomas.loc.gov/home/histdox/'
index = requests.get(base + 'fedpapers.html')
soup = BeautifulSoup.BeautifulSoup(index.text)
table = soup.find('table', {'class':'laws'})
rows = table.findAll('tr')
links = [dict(x.find('a').attrs)['href'] for x in rows[1:]]
```

```python
#Fix an anomolous link:
links[69] = 'fed_70.html'
#These are the class labels.
authors = [x.findAll('td')[2].text for x in rows[1:]]
```

```python
#Download all the papers
raw_documents = [requests.get(base+x) for x in links]
```

```python
#Parse the documents to retrieve the relevant text.
def parse_document(raw_text):
    """This gets the Federalist papers document.
    """
    soup = BeautifulSoup.BeautifulSoup(raw_text.text)
    d = soup.find('div', {'id':'main_menu1'})
    text = [x.text for x in d.findAll('p')[3:-2]]
    return ' '.join(text)
parsed_documents = [parse_document(x) for x in raw_documents]
```

The Data
---------------
Once we've dowloaded all the papers, we can begin considering how to construct numeric features from our text data. In this case we'll use word frequencies. The hypothesis behind this approach being that the different stylistic tendencies of Hamilton and Madison may be evident in their rate of use of certain words and that by identifying such words we can train a model on the known papers that will predict who was the author of the unattributed papers. You can see why this is such a popular data, it's got a labeled training set, as well as an unlabelled mystery data set to try and test methods on; it's tailor-made for classification algorithms.

```python
#Construct a frame of this data
df = pd.DataFrame({'author_class':authors, 'raw_text':parsed_documents})
#We want to get the frequency count of words in the documents.
freqs = df.raw_text.apply(lambda x: sum([nltk.FreqDist(nltk.word_tokenize(y))
                                        for y in nltk.tokenize.sent_tokenize(x)]))
total = freqs.apply(lambda x: sum(x.values()))
words = pd.concat([df, freqs.apply(pd.Series).apply(lambda x: 1000*x/total)], axis=1)
words_fill = words.fillna(0)
has_or = words_fill.author_class.str.contains('or')
words_train = words_fill[~has_or & (words_fill.author_class != 'Jay') & 
                         (~words_fill.author_class.str.contains('and'))]
words_unknown = words_fill[has_or]
```

I used nltk to construct the word-frequency counts, first tokenizing the sentences (this hopefull allows us to split on punctuation that ends a phrase, and not on periods that do not end a phrase like the ones in Mr. for example), then tokenizing those sentences into words (and punctuation). We then sum the frequencies of all the sentences into a frequency for each paper, before dividing those counts by the total number of tokens in the document and multiplying by 1000. This means that each of the columns in our training data will represent the number of times a word appears out of 1000 words. There are thousands of words represented here, but as an example of what this data looks like, see the table below.

In our training data there are 66 papers, 51 by Hamilton, and 15 by Madison. Our unattributed data has 11 papers in it.

```python
separate = ['would', 'upon']
```

```python
#Set up training data.
X = words_train[separate]
y = words_train.author_class == 'Hamilton'
```

```python
def plot_results(w):
    """Function for visualizing the results of different classifiers.
    """
    Z = w.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    figure(figsize=(10,10))
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
    ham = X[y]
    mad = X[~y]
    ham = plt.scatter(ham[[0]], ham[[1]], c='y', s=100, cmap=plt.cm.Paired, label="Hamilton")
    mad = plt.scatter(mad[[0]], mad[[1]], c='r', s=100, cmap=plt.cm.Paired, label="Madison")
    unknown = plt.scatter(words_unknown[separate][[0]],
                words_unknown[separate][[1]], s=150, marker='*',
                label='Unknown author')
    plt.xlabel('"Would" frequency')
    plt.ylabel('"Upon" frequency')
    plt.legend(handles=[ham, mad, unknown])
```

Comparing Models
===================
To compare models, I'll show some 2-d areas and color them according to how the model partitions that space for predictions. Because this is somewhat sparse data, it will be useful to look at a few troublesome points along the boundary. We'll also show how the models would have predicted the unknown texts, which previous work has mostly agreed are ALL attributable to Madison, so we'll be secretly judging our models based on how well they do according to that standard.

Logistic Regression
-------------------
Logistic regression is often the first tool anyone reaches for. It works by choosing a separating hyperplane (in this case a line) such that some error function of the predicted and actual values is minimized. In our example below, the logistic regression makes several errors on the training data, and it looks like it could be made to do better by increasing the steepness of the line it's chosen, but why doesn't it do that? It has to do with the error function used by logistic regression; each point in the space is assigned a probability by logistic regression, and the objective is optimized when those probabilities are in aggregate as close as possible to the true outcomes. Making the line steeper in this case might actually improve the classification in this example, but the model prefers to be wrong on a few points, and more sure about the remainder of the points.

Another way to think of this is that if it predicts the probability of a paper being written by Hamilton is 51%, it doesn't view that outcome much differently than if it predicts the probability as being 49% even though the class for each of those results would be different.

```python
#Fit logistic regression with normal l2 error function.
logistic = linear_model.LogisticRegression()
logistic_w = logistic.fit(X,y)
plot_results(logistic_w)
```

Support Vector Machines
-----------------------
Whereas logistic regression begins with estimating probabilities, and then derives class estimates from those, SVMs simply seek to find a way to separate the training data best. There are a couple of advantages here: first, SVMs don't get penalized for correctly classifying a result, unlike logistic regression, where every result that is less than perfectly certain results in some penalty. This means that if the data can be perfectly classified, it will be, whereas logistic regression might fail to do this. Furthermore, in the perfectly separable case, SVMs have a satisfying property that it tries to choose the best separating hyperplane, meaning that the margin of the hyperplane is maximized, this is a good property to have when you're going to use your model to subsequently predict out of sample. In cases where data is not perfectly separable, like ours, SVMs still work well, and they strike a balance between minimizing the magnitude of their errors and choosing a solution with a good margin.

In the solution below, you can see that compared to the logistic regression above, this line is made a little steeper, even though it increases the extent of misclassification of some of the training points, it provides more of a cushion between the two groups.

```python
s = svm.SVC(kernel="linear")
w_svm = s.fit(X,y)
plot_results(w_svm)
```

Gaussian Naive Bayes
====================
Gaussian Naive Bayes is an extension of Naive Bayes to allow use of continuous variables, like we have here. The assumptions of independence among the variables in Naive Bayes is still required. For example $P(upon|y=Hamilton)$ should be independent of $P(would|y=Hamilton)$. It's easy to see how this might not be true for some pairs of words that might occur together like 'in' and 'fact.' In fact, anybody who's ever made a Markov text generator has probably exploited the non-independence of individual words for a given author.

Ultimately, Naive Bayes is the technique whose assumptions are always violated but where nobody cares. Essentially, even when the independence assumption is violated, the relative probabilities are likely to be preserved so it can still be useful (though not if you want to interpret the outputs as actual probabilities. In Gausian Naive Bayes, a further assumption is required, that the features are normally distribed, which is another assumption we're straining here (especially given our frequencies are bounded at 0). 

For text-classification problems such as this, the Multinomial Naive Bayes variant is often used with word counts, but I'musing Gaussian in this case in order to use the same frequency features used for our other algorithms.

Notice, this is the first non-linear classifier we've observed (note that both SVMs and Logistic regressions can be used to create nonlinear classifiers, either via variable transformations or more advanced implementations of the algorithm). This method does ok, but it doesn't perform well with the sparse data we see here, where there is training data, the algorithm performs admirably, but in regions with few observations like along the lower part of the y-axis, it doesn't have any margin maximizing qualities like we observed with the SVM.

```python
gnb = naive_bayes.GaussianNB()
w_gnb = gnb.fit(X,y)
plot_results(w_gnb)
```

Decision Trees
===============
The decision tree below is our first perfect classifier! It classifies every element of the training and the unknown data correctly! Obviously it is the best. But no, decision trees can be made to fit data in this way very easily, and they are prone to overfitting. Notice how the area conveniently swerves around one of the more challenging points in the data? Just memorizing. The fact that it performs well out of sample in this case is basically luck.

We're not being totally fair to decision trees here by not providing a proper opportunity for cross-validation, which would be one way to avoid this kind of thing. Still, given the sparsity of our data, it's easy to see how decision trees might struggle.

```python
dtc = tree.DecisionTreeClassifier()
dtc_w = dtc.fit(X,y)
plot_results(dtc_w)
```

Random Forests
================

Random forests on the other hand have built-in overfitting avoidance, so we can see they are less reactive to single points, regardless of how pesky they are. This is nice, but since the forests are random, they can produce different results each time (try running the sell below a few times to see). They do a great job on this problem however, despite some of the challenges. In fact, it's pretty amazing to see Random Forests sometimes correctly classifies the challenging unknown paper, even though it wasn't trained on that data!

```python
rfc = ensemble.RandomForestClassifier()
rfc_w = rfc.fit(X,y)
plot_results(rfc_w)
```
